```python
#1.Program in Python to implement the DDA line algorithm->

from matplotlib import pyplot as plt 

def DDA(x0, y0, x1, y1): 

	dx = abs(x0 - x1) 
	dy = abs(y0 - y1) 

	steps = max(dx, dy) 

	xInc = dx/steps 
	yInc = dy/steps 

	x = float(x0) 
	y = float(y0) 

	x_coorinates = [] 
	y_coorinates = [] 

	for i in range(steps): 
		x_coorinates.append(x) 
		y_coorinates.append(y) 

		# increment the values 
		x = x + xInc 
		y = y + yInc 

	plt.plot(x_coorinates, y_coorinates, marker="o", 
			markersize=3) 
	plt.show() 


if __name__ == "__main__": 

	x0, y0 = 13, 53

	x1, y1 = 71, 31

	DDA(x0, y0, x1, y1) 
```
